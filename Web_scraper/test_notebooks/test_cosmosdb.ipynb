```python
from azure.cosmos import CosmosClient
from azure.cosmos import PartitionKey
import datetime
import os
```

```python
from dotenv import load_dotenv
load_dotenv('../.env')
HOST = os.getenv('HOST')
MASTER_KEY = os.getenv('KEY')
DATABASE_ID = os.getenv('DBID')
CONTAINER_ID = os.getenv('CONTAINERID')
```

```python
client = CosmosClient(HOST, {'masterKey': MASTER_KEY})
db = client.get_database_client(DATABASE_ID)
container = db.get_container_client(CONTAINER_ID)
```

```python
db.create_container_if_not_exists("Prubea", partition_key=PartitionKey(path="/productName"), indexing_policy=None, default_ttl=None, populate_query_metrics=None, offer_throughput=None, unique_key_policy=None, conflict_resolution_policy=None)
```

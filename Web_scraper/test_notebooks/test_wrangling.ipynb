```python
# The ones are with some strange ,,,,, files
M1 = pd.read_csv("mechatronics_schedules.csv")
M2 = pd.read_csv("mechatronics_schedules2.csv")
B1 = pd.read_csv("bionics_schedules.csv")
B2 = pd.read_csv("bionics_schedules2.csv")
T1 = pd.read_csv("telematics_schedules.csv")
T2 = pd.read_csv("telematics_schedules2.csv")
```

```python
M1["id"] = M1["Asignatura"] + M1["Grupo"]
M1 = M1[M1["Grupo"].notna()]
M1 = M1.sort_values("id").reset_index(drop=True)
M1
```

```python
M2["id"] = M2["Asignatura"] + M2["Grupo"]
M2 = M2.sort_values("id").reset_index(drop=True)
M2
```

```python
M = pd.concat([M1, M2], join="inner", axis=1)
M = M.loc[:,~M.columns.duplicated()]
M
```

```python
M.to_csv("Mechatronics_full.csv")
M.to_json("Mechatronics_full.json")
```

```python
M.T.to_json("Mechatronics_full_T.json")
```
